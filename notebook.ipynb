# Notebook with the simple linear model used to deploy in AWS Cloud

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
import pickle
```

## Preparing the data used by the model

```python
# Preparing the data to be used by our Linear Model
m = 150
da = 2

X = np.arange(m)
# Return a sample (or samples) from the “standard normal” distribution
epsilon = np.random.randn(m) * np.std(X)

y = da * X + epsilon
```

```python
# Plot 
plt.scatter(X, y)
plt.show()
```

## Training the simple model

```python
# Linear Regression Model
X_reshaped = X.reshape(-1, 1)
lr_model = LinearRegression()
lr_model.fit(X_reshaped, y)
```

```python
y_pred = lr_model.predict(X_reshaped)

# Plot 
plt.plot(X, y_pred, "r-", label="pred")
plt.scatter(X, y, label="true data")

plt.legend()

plt.show()
```

## Saving the model with [*pickle*](https://docs.python.org/3/library/pickle.html)

Note that *pickle* it's already within Python 3.x versions.

(Note/Warning: The pickle module is not secure. Only unpickle data you trust.)

```python
output_file = "linear_model.bin"

with open(output_file, "wb") as f_out:
    pickle.dump(lr_model, f_out)
```
